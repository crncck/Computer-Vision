```python
#!pip install matplotlib
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
picture = Image.open('../images/dog.jpg')
```

```python
type(picture)
```

```python
picture_array = np.asarray(picture)
```

```python
type(picture_array)
```

```python
picture_array.shape
```

```python
plt.imshow(picture_array)
```

```python
pic_arr_copy = picture_array.copy()
pic_arr_copy[:, :, 0]
```

0 pure black - 255 full pure red, closer to 255 more red

```python
# Red Channel Values
plt.imshow(pic_arr_copy[:, :, 0], cmap='gray') # show the image in grayscale
```

```python
# Green Channel Values 
plt.imshow(pic_arr_copy[:, :, 1], cmap='gray')
```

```python
# Blue Channel Values 
plt.imshow(pic_arr_copy[:, :, 2], cmap='gray')
```

```python
# Change values of Green channel to zero
pic_arr_copy[:, :, 1] = 0
pic_arr_copy[:, :, 1]
```

```python
# Change values of Blue channel to zero
pic_arr_copy[:, :, 2] = 0
```

```python
plt.imshow(pic_arr_copy)
```
