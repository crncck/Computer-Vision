```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('../images/dog2.jpg')
type(img)
```

Matplotlib --> Red Green Blue

OpenCV --> Blue Green Red

```python
fixed_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fixed_img)
```

```python
img_gray = cv2.imread('../images/dog2.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray, cmap='gray')
```

```python
plt.imshow(img_gray, cmap='magma')
```

### Resizing Image

```python
plt.imshow(fixed_img)
```

```python
fixed_img.shape
```

```python
new_img = cv2.resize(fixed_img, (2000, 800))
plt.imshow(new_img)
```

```python
width_ratio = 0.5 # 50% of original width
height_ratio = 0.5
```

```python
new_img = cv2.resize(fixed_img, (0,0), fixed_img, width_ratio, height_ratio)
```

```python
plt.imshow(new_img)
```

```python
new_img.shape
```

### Flipping Image

```python
new_img = cv2.flip(fixed_img, 0) # flip along the horizontal axis
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fixed_img, 1) # flip along the vertical axis
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fixed_img, -1)
plt.imshow(new_img)
```

### Saving Image

```python
type(fixed_img)
```

```python
cv2.imwrite('../images/new_dog2.jpg', fixed_img)
```

### Change Canvas Size

```python
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(111)
ax.imshow(fixed_img)
```
