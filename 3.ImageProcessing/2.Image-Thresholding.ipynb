# Image Thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('../images/rainbow.png', 0)  # 0 --> for gray scale version of image
```

```python
plt.imshow(image, cmap = 'gray')
```

## Different Treshold Types

### 1 - Binary

```python
ret, thresh1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
```

```python
ret
```

```python
plt.imshow(thresh1, cmap = 'gray')
```

### 2 - Binary Inverse

```python
ret, thresh2 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)
plt.imshow(thresh2, cmap='gray')
```

### 3 - Treshold Truncation

```python
ret, thresh2 = cv2.threshold(image, 127, 255, cv2.THRESH_TRUNC)
plt.imshow(thresh2, cmap='gray')
```

### 4 - Treshold to Zero

```python
ret, thresh2 = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO)
plt.imshow(thresh2, cmap='gray')
```

###5 - Treshold to Zero (Inverse)

```python
ret, thresh2 = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO_INV)
plt.imshow(thresh2, cmap='gray')
```

## Example

```python
image = cv2.imread("../images/crossword.jpeg", 0)
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
show_pic(image)
```

### Simple Binary

```python
ret, th1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
show_pic(th1)
```

### Adaptive Treshold

```python
th2 = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 8)
show_pic(th2)
```

```python
blended = cv2.addWeighted(src1 = th1, alpha = 0.7, src2 = th2, beta = 0.3, gamma = 0)
show_pic(blended) 
```
