# Morphological Operators

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_img():
    blank_img =np.zeros((500,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text='HELLO',org=(50,300), fontFace=font,fontScale= 5,color=(255,255,255),thickness=25,lineType=cv2.LINE_AA)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = load_img()
display_img(img)
```

## Erosion

```python
kernel = np.ones((5,5), np.uint8)
erosion = cv2.erode(img, kernel, iterations = 1)
display_img(erosion)
```

With 5 iterations

```python
img = load_img()
kernel = np.ones((5,5),np.uint8)
erosion5 = cv2.erode(img,kernel,iterations = 5)
display_img(erosion5)
```

## Opening

Useful in removing background noise

```python
img = load_img()
white_noise = np.random.randint(low = 0, high = 2, size = (500,600))
```

```python
white_noise
```

```python
white_noise = white_noise * 255
noise_img = white_noise + img
display_img(noise_img)
```

```python
opening = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernel)
display_img(opening)
```

## Closing

Useful in removing noise from foreground objects.

```python
img = load_img()
black_noise = np.random.randint(low = 0, high = 2, size = (500,600))
```

```python
black_noise
```

```python
black_noise= black_noise * -255
black_noise_img = img + black_noise
black_noise_img[black_noise_img == -255] = 0
```

```python
display_img(black_noise_img)
```

```python
closing = cv2.morphologyEx(black_noise_img, cv2.MORPH_CLOSE, kernel)
display_img(closing)
```

## Gradient

```python
img = load_img()
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
display_img(gradient)
```
