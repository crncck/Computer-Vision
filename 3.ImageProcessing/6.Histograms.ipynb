# Histograms

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
rainbow = cv2.imread('../images/rainbow.png')
show_rainbow =cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)
plt.imshow(show_rainbow)
```

```python
bricks = cv2.imread('../images/bricks.jpeg')
show_bricks = cv2.cvtColor(bricks, cv2.COLOR_BGR2RGB)
plt.imshow(show_bricks)
```

```python
hist_values = cv2.calcHist([bricks], channels=[2], mask=None, histSize=[256], ranges=[0,256])  # 2 --> red channel 
```

```python
plt.plot(hist_values)
```

```python
hist_values = cv2.calcHist([rainbow], channels=[0], mask=None, histSize=[256], ranges=[0,256])  # 0 --> blue channel
```

```python
plt.plot(hist_values)
```

```python
img = rainbow
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(histr, color = col)
    plt.xlim([0, 256])
plt.title('Rainbow Image Color Histogram')
plt.show()
```

```python
img = bricks
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(histr, color = col)
    plt.xlim([0, 256])
plt.title('Bricks Image Color Histogram')
plt.show()
```

## Histogram Equalization

```python
def display(img, cmap):
    fig = plt.figure(figsize = (10, 8))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
eagle = cv2.imread('../images/eagle.jpeg',0)
display(eagle, cmap = 'gray')
```

```python
hist_values = cv2.calcHist([eagle], channels=[0], mask=None, histSize=[256], ranges=[0,256])
```

```python
plt.plot(hist_values)
```

```python
eq_eagle = cv2.equalizeHist(eagle)
display(eq_eagle, cmap = 'gray')
```

```python
hist_values = cv2.calcHist([eq_eagle], channels=[0], mask=None, histSize=[256], ranges=[0,256])
```

```python
plt.plot(hist_values)
```

### Color Images

```python
eagle = cv2.imread('../images/eagle.jpeg')
show_eagle = cv2.cvtColor(eagle, cv2.COLOR_BGR2RGB)
hsv = cv2.cvtColor(eagle, cv2.COLOR_BGR2HSV)
```

```python
display(show_eagle, cmap = None)
```

```python
hsv[:, :, 2] = cv2.equalizeHist(hsv[:, :, 2])
```

```python
eq_eagle = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
display(eq_eagle, cmap = None)
```
