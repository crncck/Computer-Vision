# Corner Detection

```python
import cv2 
import numpy as np  
import matplotlib.pyplot as plt
```

```python
chessboard = cv2.imread('../images/chessboard.jpg')
chessboard = cv2.cvtColor(chessboard, cv2.COLOR_BGR2RGB)
plt.imshow(chessboard)
```

```python
chessboard2 = cv2.imread('../images/chessboard2.jpeg')
chessboard2 = cv2.cvtColor(chessboard2, cv2.COLOR_BGR2RGB)
plt.imshow(chessboard2)
```

## Harris Corner Detection

```python
gray_chess1 = cv2.cvtColor(chessboard, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray_chess1)
```

```python
dst = cv2.cornerHarris(src = gray, blockSize = 2, ksize = 3, k = 0.04)
```

```python
dst = cv2.dilate(dst, None)
```

```python
chessboard[dst > 0.01 * dst.max()] = [255, 0, 0]
plt.imshow(chessboard)
```

```python
gray_chess2 = cv2.cvtColor(chessboard2, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray_chess2)
```

```python
dst = cv2.cornerHarris(src = gray, blockSize = 2, ksize = 3, k = 0.04)
```

```python
dst = cv2.dilate(dst, None)
```

```python
chessboard2[dst > 0.01 * dst.max()] = [255, 0, 0]
plt.imshow(chessboard2)
```

## Shi-Tomasi Corner Detection

```python
chessboard = cv2.imread('../images/chessboard.jpg')
chessboard = cv2.cvtColor(chessboard, cv2.COLOR_BGR2RGB)
gray_chessboard = cv2.cvtColor(chessboard, cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(gray_chessboard, 64, 0.01, 10)
corners = np.int0(corners)

for i in corners:
    x, y = i.ravel()
    cv2.circle(chessboard,(x, y), 3, 255, -1)

plt.imshow(chessboard)
```

```python
chessboard2 = cv2.imread('../images/chessboard2.jpeg')
chessboard2 = cv2.cvtColor(chessboard2, cv2.COLOR_BGR2RGB)
gray_chessboard = cv2.cvtColor(chessboard2, cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(gray_chessboard, 80, 0.01, 10)
corners = np.int0(corners)

for i in corners:
    x, y = i.ravel()
    cv2.circle(chessboard2,(x, y), 3, 255, -1)

plt.imshow(chessboard2)
```
